```python
# import libraries
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
# load in our data
data = pd.read_csv("insur_arrests.csv")
```

```python
# view it
data
```

```python
# replace all the line breaks with semicolons -- for r for jenna
data["charges_clean"] = data["charges"].str.replace('\n\n', '; ')
# replace all the \n with semicolons, too -- for r for jenna
data["charges_clean"] = data["charges_clean"].str.replace('\n', '; ')

# now, replace the weird \xa0, with a single space, which is basically code for single space -- more about it here:
# https://stackoverflow.com/questions/10993612/how-to-remove-xa0-from-string-in-python
data["charges_clean"] = data["charges_clean"].str.replace('\xa0', ' ')

# split the cells on semi colon to get a list for analysis
data["charges_clean_list"] = data["charges_clean"].str.split('; ')

data
```

```python
# view the charges

# initialize empty list charge_list for all cahrges
charges_list = []

# initialize another empty list for clean charges (lowercased)
clean_list = []

# for row in df
for index,row in data.iterrows():
    # for each item in charges_clean_list
    for list_item in row["charges_clean_list"]:
        # add the item to our charges_list
        if list_item not in charges_list:
            charges_list.append(list_item)
            
# lowercase the charges_list
for crime in charges_list:
    clean_list.append(crime.lower())

# see how many unique items in clean_list
print(len(clean_list))
print('---')

# view clean_list
print(clean_list)
```
